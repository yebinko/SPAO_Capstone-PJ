```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('22년매장재고_수정.csv')
```

```python
list = ['기간판매량','잔여재고량', '누적판매율[입고량대비]', '기간판매율[입고량대비]']
for i in list:
    # '기간판매량' 열을 float으로 변환하고, 'NaN' 값을 null로 처리합니다.
    df[i] = pd.to_numeric(df[i], errors='coerce')

    # 'NaN' 값을 null로 처리하려면 np.nan을 사용합니다.
    df[i].replace('NaN', np.nan, inplace=True)
```
