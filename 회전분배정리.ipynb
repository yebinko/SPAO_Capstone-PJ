```python
def get_stdev(df):
    df2 = df[df['판매주차수'] == 18]
    df3 = df2.groupby(['매장코드', '시즌']).mean()
    df3.reset_index(inplace=True)
    df3.dropna(inplace=True)
    print(df3['소진율'].describe())
    df_season1 = df3[df3['시즌'] == 1]
    df_season2 = df3[df3['시즌'] == 2]
    df_season3 = df3[df3['시즌'] == 3]
    df_season4 = df3[df3['시즌'] == 4]
    range_season1 = df_season1['소진율'].max() -df_season1['소진율'].min()
    range_season2 = df_season2['소진율'].max() -df_season2['소진율'].min()
    range_season3 = df_season3['소진율'].max() -df_season3['소진율'].min()
    range_season4 = df_season4['소진율'].max() -df_season4['소진율'].min()
    stdev_season1 = df_season1['소진율'].std()
    stdev_season2 = df_season2['소진율'].std()
    stdev_season3 = df_season3['소진율'].std()
    stdev_season4 = df_season4['소진율'].std()
    return stdev_season1, stdev_season2, stdev_season3, stdev_season4
```

# 회전분배가 유의미했던 % (매장, 스타일코드)

```python
def spin(df):
    df21_2 = df[(df['판매주차수'] == 19) |(df['판매주차수'] == 20) 
                |(df['판매주차수'] == 21) |(df['판매주차수'] == 22) |(df['판매주차수'] == 23)|(df['판매주차수'] == 24)
                |(df['판매주차수'] == 25)|(df['판매주차수'] == 26) |(df['판매주차수'] == 27) |(df['판매주차수'] == 28)
                |(df['판매주차수'] == 29)|(df['판매주차수'] == 30) |(df['판매주차수'] == 31) |(df['판매주차수'] == 32)
                |(df['판매주차수'] == 33)|(df['판매주차수'] == 34) |(df['판매주차수'] == 35) |(df['판매주차수'] == 36)
                |(df['판매주차수'] == 37)|(df['판매주차수'] == 38)]
    df21_18 = df[df['판매주차수'] == 18]
    df21_21 = df21_2.groupby(['매장코드', '시즌', '스타일코드']).sum()
    df21_18_21 = df21_18.groupby(['매장코드','시즌', '스타일코드']).sum()
    df21_21.reset_index(inplace=True)
    df21_18_21.reset_index(inplace=True)
    df_merge = pd.merge(df21_21, df21_18_21[['매장코드', '스타일코드', '잔여재고량']], on=['매장코드', '스타일코드'])
    df_merge2 = df_merge[df_merge['기간입고량']!=0]
    df_merge2['기간판매량/기간입고량(회전분배량)'] = df_merge2['기간판매량'] / (df_merge2['기간입고량']+ 0.001)
    df_merge2['기간판매량/잔여재고량'] = df_merge2['기간판매량'] / (df_merge2['잔여재고량_y']+ 0.001)
    df_merge2['기간판매량/합'] = df_merge2['기간판매량'] / (df_merge2['기간입고량'] + df_merge2['잔여재고량_y']+ 0.001)
    df_merge3 = df_merge2[df_merge2['기간판매량/잔여재고량']>1]
    print('전체: ', len(df_merge3) /len(df_merge2))
    df_merge2_season = df_merge2.groupby('시즌').count()
    df_merge3_season = df_merge3.groupby('시즌').count()
    df_merge2_store = df_merge2.groupby('매장코드').count()
    df_merge3_store = df_merge3.groupby('매장코드').count()
    df_merge2_store.reset_index(inplace=True)
    df_merge3_store.reset_index(inplace=True)
    print('시즌별: ', df_merge3_season['기간판매량/잔여재고량'] /df_merge2_season['기간판매량/잔여재고량'])
    print('매장별: ', df_merge3_store['기간판매량/잔여재고량'] / df_merge2_store['기간판매량/잔여재고량'])
    return df_merge3_store
```
