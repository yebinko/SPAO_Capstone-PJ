# 1.결품률

```python
def cal_minus(x):
    minus = len(df[(df['스타일코드'] == x) & (df['잔여재고량'] <= 0)]) / len(df[df['스타일코드'] == x]) * 100
    name = df[df['스타일코드']== x]['스타일명'].unique()
    period = df[df['스타일코드'] == x]['Date'].iloc[0]
    return x, name, period, minus
```

```python
codes = []
minuses = []
names = []
periods = []
for x in df['스타일코드'].unique():
    minus = len(df[(df['스타일코드'] == x) & (df['잔여재고량'] <= 0)]) / len(df[df['스타일코드'] == x]) * 100
    name = df[df['스타일코드']== x]['스타일명'].unique()
    period = df[df['스타일코드'] == x]['Date'].iloc[0]
    codes.append(x)
    minuses.append(minus)
    names.append(name)
    periods.append(period)
```

```python
df_minus = pd.DataFrame({'스타일코드':codes, '스타일명':names, '판매시작일':periods, '결품률':minuses})
#df_minus.to_csv('기업공유_결품률.csv', encoding='utf-8-sig')
df_minus
```

```python
df2 = df[df['스타일코드'] != 'SPTAE11W05']
df2 = df2[df2['스타일코드'] != 'SPTAE11W06']
```

```python
def get_pj_result(df, table, grade):
        
        forman_a = table[table['포맨'] == grade]['매장코드'].unique()
        df = df[df['기간판매율[입고량대비]'] > 0]
        df_forman_a = df[df['매장코드'].isin(forman_a)]
        df_forman_a['아이템'] = df_forman_a['스타일코드'].str[2:4]
        df_forman_a2 = df_forman_a[df_forman_a['아이템'] == 'TA']
        df_forman_a3 = df_forman_a2[df_forman_a2['스타일코드'] != 'SPTAE23M07']

        codes = []
        minuses = []
        names = []
        periods = []
        for x in df_forman_a3['스타일코드'].unique():
            a = len(df_forman_a3[(df_forman_a3['스타일코드'] == x) & (df_forman_a3['잔여재고량'] <= 0)])
            b = len(df_forman_a3[df_forman_a3['스타일코드'] == x])
            minus = ( (a/b)  * 100 )
            name = df_forman_a3[df_forman_a3['스타일코드']== x]['스타일명'].unique()
            period = df_forman_a3[df_forman_a3['스타일코드'] == x]['Date'].iloc[0]
            codes.append(x)
            minuses.append(minus)
            names.append(name)
            periods.append(period)

        df_minus_a = pd.DataFrame({'스타일코드':codes, '스타일명':names, '판매시작일':periods, '결품률':minuses})

        # df_early = df_forman_a3[(df_forman_a3['스타일코드'] == 'SPTAE11W05') 
        #                         | (df_forman_a3['스타일코드'] == 'SPTAE11W06')]
        # df_early2 = df_early[df_early['Date'] <= 240106]
        # df_mean_early = df_early2.groupby(['매장코드', '매장명', '스타일코드', 'Date'])[['기간판매율[입고량대비]', '누적판매율[입고량대비]']].mean()
        # df_mean_early2 = df_mean_early.groupby(['스타일코드', 'Date'])['기간판매율[입고량대비]'].std()
        # df_mean_early3 = df_mean_early2.groupby('스타일코드').mean()
        # df_mean_early3 = pd.DataFrame(df_mean_early3).reset_index()
        # df_mean_early3['복종'] = df_mean_early3['스타일코드'].str[7]

        # df_forman_a3 = df_forman_a3[df_forman_a3['스타일코드'] != 'SPTAE11W05']
        # df_forman_a3 = df_forman_a3[df_forman_a3['스타일코드'] != 'SPTAE11W06']
        df_mean_a = df_forman_a3.groupby(['매장코드', '매장명', '스타일코드', 'Date'])[['기간판매율[입고량대비]', '누적판매율[입고량대비]']].mean()
        df_mean2_a = df_mean_a.groupby(['스타일코드', 'Date'])['기간판매율[입고량대비]'].std()
        df_mean3_a = df_mean2_a.groupby('스타일코드').mean()
        df_mean3_a = pd.DataFrame(df_mean3_a).reset_index()
        df_mean3_a['복종'] = df_mean3_a['스타일코드'].str[7]

        # df_mean3_a = pd.concat([df_mean3_a, df_mean_early3], axis=0)

        df_final_a = pd.merge(df_minus_a, df_mean3_a, on='스타일코드')
        df_final_a.columns = ['스타일코드', '스타일명', '판매시작일', '결품률','매장간 소진율 편차', '복종']

        pj_code = ['SPTAEA9W01', 'SPTAE23M03', 'SPTAE23M04']

        pj = []
        selected_rows = df_final_a[df_final_a['스타일코드'].isin(pj_code)]
        pj_minus_mean = selected_rows['결품률'].mean()
        pj_sojin_mean = selected_rows['매장간 소진율 편차'].mean()
        pj.append(pj_minus_mean)
        pj.append(pj_sojin_mean)

        npj = []

        not_selected_rows = df_final_a[~df_final_a['스타일코드'].isin(pj_code)]
        npj_minus_mean = not_selected_rows['결품률'].mean()
        npj_sojin_mean = not_selected_rows['매장간 소진율 편차'].mean()
        npj.append(npj_minus_mean)
        npj.append(npj_sojin_mean)

        

        return pd.DataFrame({'파일럿 적용':pj, '파일럿 미적용': npj})
```

```python
def get_grade_sojin(df, table, grade):
    df = df[df['기간판매율[입고량대비]'] > 0]
    forman_a = table[table['포맨'] == grade]['매장코드'].unique()
    df_forman_a = df[df['매장코드'].isin(forman_a)]
    df_forman_a['아이템'] = df_forman_a['스타일코드'].str[2:4]
    df_forman_a2 = df_forman_a[df_forman_a['아이템'] == 'TA']
    df_forman_a3 = df_forman_a2[df_forman_a2['스타일코드'] != 'SPTAE23M07']

    df_mean_a = df_forman_a3.groupby(['매장코드', '매장명', '스타일코드', 'Date'])[['기간판매율[입고량대비]', '누적판매율[입고량대비]']].mean()
    df_mean_a2 = df_mean_a.groupby(['스타일코드', 'Date'])['기간판매율[입고량대비]'].mean()
    df_mean_a2 = pd.DataFrame(df_mean_a2)
    df_mean_a2 = df_mean_a2.reset_index()
    df_mean_a2.columns = ['스타일코드', 'Date', f'{grade}등급매장의 평균소진율']

    return df_mean_a2
```

```python
grade_a = get_grade_sojin(df, table, 'A')
grade_b = get_grade_sojin(df, table, 'B')
grade_c = get_grade_sojin(df, table, 'C')
grade_d = get_grade_sojin(df, table, 'D')
grade_e = get_grade_sojin(df, table, 'E')
grade_f = get_grade_sojin(df, table, 'F')
```

```python
grade_df = pd.merge(grade_a, grade_b, on=['스타일코드', 'Date'], how='outer')
grade_df = pd.merge(grade_df, grade_c, on=['스타일코드', 'Date'], how='outer')
grade_df = pd.merge(grade_df, grade_d, on=['스타일코드', 'Date'], how='outer')
grade_df = pd.merge(grade_df, grade_e, on=['스타일코드', 'Date'], how='outer')
grade_df = pd.merge(grade_df, grade_f, on=['스타일코드', 'Date'], how='outer')
grade_df
```
