```python
import pandas as pd
```

```python
df = pd.read_csv('cctv data.csv')
```

```python
df
```

```python
df['성별인원'] = df['9세 이하'] + df['10대'] + df['20대'] + df['30대']+df['40대']+df['50대']+df['60세 이상']
```

```python
df_all = df.groupby('매장').sum()
```

```python
df_age = df.groupby('매장').sum()
```

```python
df_all.columns
```

```python
df_gender = df.drop(['9세 이하', '10대', '20대', '30대', '40대', '50대', '60세 이상'], axis=1)
df_gender
```

```python
df_pivoted = df_gender.pivot(index='매장', columns='성별', values='성별인원').reset_index()
df_pivoted.columns.name = None  # 컬럼 이름 제거

# 열 이름 변경
df_pivoted.columns = ['매장코드', '남성인원', '여성인원']
```

```python
df_pivoted
```

```python
df_age.to_csv('연령별입점고객.csv', encoding='utf-8-sig')
```

```python
df_age = pd.read_csv('연령별입점고객.csv')
```

```python
df_age
```

```python
df_all = pd.merge(df_age, df_pivoted, right_on='매장코드', left_on='매장')
```

```python
df_all['총인원'] = df_all['여성인원'] + df_all['남성인원']
```

```python
df_all
```

```python
df_all.columns
```

```python
for i in ['9세 이하', '10대', '20대', '30대', '40대', '50대', '60세 이상', '남성인원', '여성인원']:
    df_all[f'{i} 비율'] = df_all[i]/df_all['총인원']
```

```python
df_all.to_csv('입점고객_리그레션용.csv', encoding='utf-8-sig')
```
