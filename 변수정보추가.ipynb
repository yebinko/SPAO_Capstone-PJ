```python
import pandas as pd
```

```python
nam = pd.read_csv('남캐_두시즌_고객x.csv', index_col=0)
```

```python
yeo = pd.read_csv('남캐_두시즌_고객x.csv', index_col=0)
fm = pd.read_csv('여캐_두시즌_고객x.csv', index_col=0)
fw = pd.read_csv('포맨_두시즌_고객x.csv', index_col=0)
be = pd.read_csv('포우먼_두시즌_고객x.csv', index_col=0)
jap = pd.read_csv('잡화_두시즌_고객x.csv', index_col=0)
```

```python
var_man = pd.read_excel('1020 1차 분배등급_공유.xlsx', index_col=0)
var_yeo = pd.read_excel('1020 1차 분배등급_공유.xlsx',sheet_name='여캐',  index_col=0)
var_fm = pd.read_excel('1020 1차 분배등급_공유.xlsx',sheet_name='포맨',  index_col=0)
var_fw = pd.read_excel('1020 1차 분배등급_공유.xlsx',sheet_name='포우먼',  index_col=0)
var_be = pd.read_excel('1020 1차 분배등급_공유.xlsx', sheet_name='베이직콜라보', index_col=0)
var_jap = pd.read_excel('1020 1차 분배등급_공유.xlsx', sheet_name='잡화', index_col=0)
```

```python
def get_final(df1, df2):
    df_final = pd.merge(df1, df2, left_on=['y_year','매장코드'], right_on=['year', '매장코드'], how='outer')
    df_final = df_final.sort_values('year')
    return df_final
```

```python
nam_final = get_final(nam, var_man)
yeo_final = get_final(yeo, var_yeo)
fm_final = get_final(fm, var_fm)
fw_final = get_final(fw, var_fw)
be_final = get_final(be, var_be)
jap_final = get_final(jap, var_jap)
```

```python
nam_final.to_excel('남캐변수별정보포함.xlsx')
```
